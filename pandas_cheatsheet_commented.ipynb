# üì¶ Pandas Cheat Sheet (Series & DataFrame)

## ‚úÖ Import

```python
import pandas as pd      # Import biblioteki pandas
import numpy as np       # Import biblioteki numpy
```

## üìö Tworzenie danych

```python
s = pd.Series([1, 2, 3], name="example")    # Tworzy Series z nazwƒÖ
df = pd.DataFrame({                         # Tworzy DataFrame ze s≈Çownika
    'A': [1, 2, 3],
    'B': ['x', 'y', 'z']
})
```

## üß≠ Eksploracja

```python
df.head()          # Pierwsze 5 wierszy
df.tail()          # Ostatnie 5 wierszy
df.shape           # Liczba wierszy i kolumn (krotka)
df.info()          # Informacje o kolumnach i typach danych
df.describe()      # Statystyki dla kolumn liczbowych
df.columns         # Lista nazw kolumn
df.index           # Indeksy wierszy
df.dtypes          # Typy danych kolumn
```

## üîç Selekcja danych

```python
df['A']                # Zwraca kolumnƒô A jako Series
df[['A', 'B']]         # Zwraca wybrane kolumny jako DataFrame
df.iloc[0]             # Wiersz na pozycji 0 (wg indeksu numerycznego)
df.loc[0]              # Wiersz o indeksie 0 (wg etykiety)
df.iloc[0, 1]          # Warto≈õƒá z 0. wiersza i 1. kolumny
df.loc[0, 'A']         # Warto≈õƒá z 0. wiersza i kolumny A
```

## üßπ Filtrowanie

```python
df[df['A'] > 1]                                 # Wiersze gdzie A > 1
df[(df['A'] > 1) & (df['B'] == 'y')]            # Wiersze gdzie A > 1 i B == 'y'
```

## ‚úçÔ∏è Modyfikacja

```python
df['C'] = df['A'] * 2                           # Nowa kolumna C jako A * 2
df['B'] = df['B'].str.upper()                   # Zamiana tekstu w kolumnie B na wielkie litery
df.drop('C', axis=1)                            # Usuniƒôcie kolumny C
df.drop(0, axis=0)                              # Usuniƒôcie wiersza o indeksie 0
df.rename(columns={'A': 'Alpha'})               # Zmiana nazwy kolumny A na Alpha
```

## üîÅ Funkcje

```python
df.apply(np.sqrt)                              # Zastosowanie pierwiastka do ka≈ºdego elementu
df['A'].apply(lambda x: x**2)                  # Zastosowanie lambda do kolumny A
```

## üìä Grupowanie

```python
df.groupby('B').mean()                         # ≈örednia warto≈õci w grupach B
df.groupby('B')['A'].sum()                     # Suma A w grupach B
```

## üßΩ Czyszczenie

```python
df.isnull()                                    # Sprawdza brakujƒÖce warto≈õci
df.dropna()                                    # Usuwa wiersze z brakujƒÖcymi warto≈õciami
df.fillna(0)                                   # Zastƒôpuje NaN zerami
df.duplicated()                                # Sprawdza duplikaty wierszy
df.drop_duplicates()                           # Usuwa duplikaty wierszy
```

## üîÄ Sortowanie

```python
df.sort_values('A')                            # Sortuje po kolumnie A rosnƒÖco
df.sort_values(['A', 'B'])                     # Sortuje po A, potem B
```

## üì§ Wczytywanie / zapisywanie

```python
pd.read_csv('plik.csv')                        # Wczytuje plik CSV do DataFrame
df.to_csv('plik.csv', index=False)             # Zapisuje DataFrame do CSV bez indeksu
```

## üß™ Indeksy

```python
df.set_index('A')                              # Ustawia kolumnƒô A jako indeks
df.reset_index()                               # Resetuje indeks do domy≈õlnego
```

## üîé Szybkie analizy

```python
df['A'].value_counts()                         # Liczba wystƒÖpie≈Ñ unikalnych warto≈õci
df['A'].unique()                               # Lista unikalnych warto≈õci
df['A'].nunique()                              # Liczba unikalnych warto≈õci
```



# üìà Pandas / Matplotlib Plotting Cheat Sheet

## ‚úÖ Import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## üêº Wbudowane wykresy w Pandas

```python
df['A'].plot()                         	# liniowy
df.plot(kind='bar')                   	# s≈Çupkowy
df.plot(kind='hist')                  	# histogram
df.plot(kind='box')                   	# boxplot
df.plot(kind='scatter', x='A', y='B') 	# scatter
df.plot(kind='pie', y='A')            	# wykres ko≈Çowy
```

## ‚öôÔ∏è Opcje formatowania

```python
df.plot(title='Wykres', figsize=(8, 4), grid=True)
```

## üé® Matplotlib ‚Äî kontrola totalna

```python
plt.plot(df['A'])
plt.scatter(df['A'], df['B'])
plt.bar(df['A'], df['B'])
plt.hist(df['A'], bins=10)
plt.boxplot(df['A'])
plt.title("Tytu≈Ç")
plt.xlabel("O≈õ X")
plt.ylabel("O≈õ Y")
plt.grid(True)
plt.show()
```
