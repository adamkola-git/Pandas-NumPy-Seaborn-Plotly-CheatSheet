# üì¶ Pandas Cheat Sheet (Series & DataFrame)

### ‚úÖ Import

```python
import pandas as pd
import numpy as np
```

## üìö Tworzenie danych

```python
s = pd.Series([1, 2, 3], name="example")
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': ['x', 'y', 'z']
})
```

## üß≠ Eksploracja

```python
df.head()         
df.tail()         
df.shape          
df.info()         
df.describe()     
df.columns        
df.index          
df.dtypes         
```

## üîç Selekcja danych

```python
df['A']
df[['A', 'B']]
df.iloc[0]
df.loc[0]
df.iloc[0, 1]
df.loc[0, 'A']
```

## üßπ Filtrowanie

```python
df[df['A'] > 1]
df[(df['A'] > 1) & (df['B'] == 'y')]
```

## ‚úçÔ∏è Modyfikacja

```python
df['C'] = df['A'] * 2
df['B'] = df['B'].str.upper()
df.drop('C', axis=1)
df.drop(0, axis=0)
df.rename(columns={'A': 'Alpha'})
```

## üîÅ Funkcje

```python
df.apply(np.sqrt)
df['A'].apply(lambda x: x**2)
```

## üìä Grupowanie

```python
df.groupby('B').mean()
df.groupby('B')['A'].sum()
```

## üßΩ Czyszczenie

```python
df.isnull()
df.dropna()
df.fillna(0)
df.duplicated()
df.drop_duplicates()
```

## üîÄ Sortowanie

```python
df.sort_values('A')
df.sort_values(['A', 'B'])
```

## üì§ Wczytywanie / zapisywanie

```python
pd.read_csv('plik.csv')
df.to_csv('plik.csv', index=False)
```

## üß™ Indeksy

```python
df.set_index('A')
df.reset_index()
```

## üîé Szybkie analizy

```python
df['A'].value_counts()
df['A'].unique()
df['A'].nunique()
```



# üìà Pandas / Matplotlib Plotting Cheat Sheet

## ‚úÖ Import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## üêº Wbudowane wykresy w Pandas

```python
df['A'].plot()                         # liniowy
df.plot(kind='bar')                   # s≈Çupkowy
df.plot(kind='hist')                  # histogram
df.plot(kind='box')                   # boxplot
df.plot(kind='scatter', x='A', y='B') # scatter
df.plot(kind='pie', y='A')            # wykres ko≈Çowy
```

## ‚öôÔ∏è Opcje formatowania

```python
df.plot(title='Wykres', figsize=(8, 4), grid=True)
```

## üé® Matplotlib ‚Äî kontrola totalna

```python
plt.plot(df['A'])
plt.scatter(df['A'], df['B'])
plt.bar(df['A'], df['B'])
plt.hist(df['A'], bins=10)
plt.boxplot(df['A'])
plt.title("Tytu≈Ç")
plt.xlabel("O≈õ X")
plt.ylabel("O≈õ Y")
plt.grid(True)
plt.show()
```
